```python
! pip install graphviz
import numpy as np 
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
from keras.datasets import mnist
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import train_test_split
from sklearn.decomposition import PCA
from sklearn import tree
from sklearn.metrics import classification_report
```

### Loading the Mnist dataset from Keras.datasets

```python
(dataset_X, dataset_y), (test_X, test_y) = mnist.load_data()
```

### Exploring the dataset

```python
for i in range(9):
    plt.subplot(330 + 1 + i)
    plt.imshow(dataset_X[i], cmap=plt.get_cmap('gray'))
plt.tight_layout(pad=0.5)
plt.show()
```

### Flattening the 28x28 matrix's into 784 dim-array

```python
X = dataset_X.reshape((dataset_X.shape[0], 28*28))
Y=dataset_y
testX = test_X.reshape((test_X.shape[0], 28* 28))
testY=test_y
```

### Splitting the data into Train and test sets

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.2, random_state = 0)
```

### PLotting Histogram to check that Train data set is well balanced

```python
fig, ax = plt.subplots()
plt.xlabel("Labels")
plt.ylabel("Number of entries label")
ax.hist(y_train, bins = 20)
plt.show()
```

### Performing PCA ; capture 95% variation

```python
pca = PCA(0.97)

X_train = pca.fit_transform(X_train)
X_test = pca.transform(X_test)
testX=pca.transform(testX)

explained_variance = pca.explained_variance_ratio_
```

### Function to calculate Recall, Precission, F1 score

```python
def caculate_Recall(matr):
    matr=np.array(matr)
    colsum=matr.sum(axis=1)
    l=list()
    for i in range(np.shape(matr)[0]):
        l.append(matr[i][i]/colsum[i])
    return l
```

```python
def caculate_Precision(matr):
    matr=np.array(matr)
    colsum=matr.sum(axis=0)
    l=list()
    for i in range(np.shape(matr)[0]):
        l.append(matr[i][i]/colsum[i])
    return l
```

```python
def cal_F1(matr):
    r=caculate_Recall(matr)
    p=caculate_Precision(matr)
    f=list()
    for i in range(len(p)):
        f.append(2*r[i]*p[i]/(r[i]+p[i]))
    return f
```

```python
def cal_weighted(matr,y_true):
        l=[0]*10
        for i in y_test:
            l[i]+=1
        f1=cal_F1(matr)
        r=caculate_Recall(matr)
        p=caculate_Precision(matr)
        averageF1=np.sum(np.multiply(f1,l))/len(y_true)
        averagep=np.sum(np.multiply(p,l))/len(y_true)
        averager=np.sum(np.multiply(r,l))/len(y_true)
        return [averageF1,averagep,averager]
```

### Code to build one Classification Tree of Height 6

```python
clf = tree.DecisionTreeClassifier(random_state=0, max_depth=6)
clf = clf.fit(X_train, y_train)
```

```python
y_pred=clf.predict(X_test)
```

### Print the confusion matrix

```python
conf_matr=confusion_matrix(y_test, y_pred)
```

```python
conf_matr
```

### Making a series of models from height 5-24 to get the best height of tree

```python
models=list()
for i in range(5,25):
    clf = tree.DecisionTreeClassifier(random_state=0, max_depth=i)
    clf = clf.fit(X_train, y_train)
    y_pred=clf.predict(X_test)
    conf_matr=confusion_matrix(y_test, y_pred)
    f1_p_r=cal_weighted(conf_matr,y_test)
    dic=dict()
    dic["model"]=clf
    dic['pred']=y_pred
    dic["conf"]=conf_matr
    dic['F1']=f1_p_r[0]
    dic["Prec"]=f1_p_r[1]
    dic["Recall"]=f1_p_r[2]
    dic["Height"]=i
    models.append(dic)
```

```python
f1=list()
p=list()
r=list()
for i in range(len(models)):
    f1.append(models[i]['F1'])
    p.append(models[i]["Prec"])
    r.append(models[i]['Recall'])
```

### Plot the variation of diffrent metrics Vs height of tree

```python
x=range(5,25)
figure, axis = plt.subplots(1, 3)
axis[2].plot(x, f1)
axis[2].set_title("F1 score Vs Height of Tree")
axis[1].plot(x, p)
axis[1].set_title("Precision Vs Height of Tree")
axis[0].plot(x, r)
axis[0].set_title("Recall Vs Height of Tree")
plt.subplots_adjust(left=0.3,
                    bottom=0.1, 
                    right=2, 
                    top=0.9, 
                    wspace=0.5, 
                    hspace=0.5)

plt.show()
```

### Select the model with best F1 score

```python
max_value=max(f1)
model_selected=models[f1.index(max_value)]
```

### Predicting the train data

```python
y_pred=model_selected["model"].predict(testX)
```

### Report on performance of model

```python
print(classification_report(test_y, y_pred))
```

### Confusion matrix and optimum height

```python
conf_matr=confusion_matrix(test_y, y_pred)
print(conf_matr)
print("Height of tree ",model_selected["Height"])
```
